```python
from pykeen.datasets import CoDExSmall
```

```python
dataset = CoDExSmall(create_inverse_triples=True)
```

```python
import pykeen
pykeen.env()
```

## NodePiece

```python
from pykeen.pipeline import pipeline
from pykeen.models import NodePiece

result = pipeline(
    dataset=dataset,
    model=NodePiece,
    model_kwargs=dict(
        tokenizers=["AnchorTokenizer", "RelationTokenizer"],
        num_tokens=[20, 6],
        tokenizers_kwargs=[
            dict(
                selection="RandomAnchorSelection",
                selection_kwargs=dict(
                    num_anchors=200,
                ),
                searcher="ScipySparse",
            ),
            dict(),  # empty dict for the RelationTokenizer - it doesn't need any kwargs
        ],
        embedding_dim=200,
        interaction="rotate",
    ),
    training_kwargs=dict(num_epochs=400),
    device='cuda',
    stopper = 'early'
)
```

```python
result.metric_results.get_metric("inverseharmonicmeanrank")
```

```python
result.save_to_directory('doctests/Model_Node_Piece_Data_CoDExSmall')
```

## ComplEx

```python
from pykeen.pipeline import pipeline
from pykeen.models import ComplEx

result = pipeline(
    dataset=dataset,
    model=ComplEx,
    model_kwargs=dict(embedding_dim = 512, entity_initializer="xavier_uniform", relation_initializer="xavier_uniform"),
    training_kwargs=dict(num_epochs=400, batch_size=1024, sampler='1vsall'),
    device='cuda',
    optimizer = 'adam',
    optimizer_kwargs = dict(lr=0.02),
    training_loop = 'slcwa',
    loss='crossentropy',
    stopper = 'early'

)
```

```python
result.metric_results.get_metric("inverseharmonicmeanrank")
```

```python
result.metric_results.to_df().to_csv("evaluation13.csv")
```

```python
from pykeen.predict import predict_all
pack = predict_all(model = result.model, k = 40000)
```

```python
predictionsDF = pack.process().df
```

```python
entity_to_id = dataset.entity_to_id
id_to_entity = {value: key for key, value in entity_to_id.items()}
relation_to_id = dataset.relation_to_id
id_to_relation = {value: key for key, value in relation_to_id.items()}
```

```python
import requests
def fetch_label_from_wikidata(wikidata_id):
    # Prepare the Wikidata API URL
    url = f'https://www.wikidata.org/w/api.php?action=wbgetentities&ids={wikidata_id}&props=labels&format=json'
    # Send a GET request to the Wikidata API
    response = requests.get(url)
    # Check if the request was successful
    if response.status_code == requests.codes.ok:
        data = response.json()
        # Extract the label from the response
        try:
            label = data['entities'][wikidata_id]['labels']['en']['value']
            return label
        except KeyError:
            return f"No label found for entity ID {wikidata_id}"
    else:
        return "Request failed"
```

```python
import pandas as pd
import numpy as np
predictionsDF = pd.read_csv('allPredictionsComplExOnCoDExSmall_dataset_id.csv')
# Replace ids from dataset with their values in the dataframe
predictionsDF.head_id = [id_to_entity[id] for id in predictionsDF.head_id]
predictionsDF.relation_id = [id_to_relation[id] for id in predictionsDF.relation_id]
predictionsDF.tail_id = [id_to_entity[id] for id in predictionsDF.tail_id]

predictionsDF.to_csv("allPredictionsComplExOnCoDExSmall_wikidata_id.csv", index = None)

# Filter rows not in the dataset
predictions = predictionsDF[['head_id', 'relation_id', 'tail_id']].values
predictions = [tuple(row) for row in predictions.tolist()]
dataset_triples_set = set(map(tuple, dataset.training.triples))

include_prediction_arr = []
for prediction in predictions:
  include_prediction = !(prediction[0] == prediction[2]) && !(prediction in dataset_triples_set)
  include_prediction_arr.append(include_prediction)

new_links_df = predictionsDF[include_prediction_arr]
new_links_df.to_csv("filteredPredictionsComplExOnCoDExSmall_wikidata_id.csv", index = None)

# Get all unique wikidate ids
wikidata_ids = list(set(new_links_df.head_id.tolist() + new_links_df.tail_id.tolist() + new_links_df.relation_id.tolist()))
# Fetch all labels from wikidata and save them in a dictionary
wikidata_id_to_label = {}
for wikidata_id in wikidata_ids:
    label = fetch_label_from_wikidata(wikidata_id)
    wikidata_id_to_label[wikidata_id] = label
# Replace ids from wikidata with their labels in the dataframe
new_links_df.head_id = [wikidata_id_to_label[id] for id in new_links_df.head_id]
new_links_df.relation_id = [wikidata_id_to_label[id] for id in new_links_df.relation_id]
new_links_df.tail_id = [wikidata_id_to_label[id] for id in new_links_df.tail_id]
# Save the dataframe with the labels in a CSV
new_links_df.to_csv("filteredPredictionsComplExOnCoDExSmall_wikidata_label.csv", index = None)
```

```python
from pykeen.datasets import CoDExMedium
```

```python
dataset = CoDExMedium(create_inverse_triples=True)
```

```python
import pykeen
pykeen.env()
```

```python
from pykeen.pipeline import pipeline
from pykeen.models import NodePiece

result = pipeline(
    dataset=dataset,
    model=NodePiece,
    model_kwargs=dict(
        tokenizers=["AnchorTokenizer", "RelationTokenizer"],
        num_tokens=[20, 6],
        tokenizers_kwargs=[
            dict(
                selection="RandomAnchorSelection",
                selection_kwargs=dict(
                    num_anchors=1500,
                ),
                searcher="ScipySparse",
            ),
            dict(),  # empty dict for the RelationTokenizer - it doesn't need any kwargs
        ],
        embedding_dim=100,
        interaction="rotate",
    ),
    stopper='early',
    device='cuda',
    training_loop = "lcwa",
    training_kwargs=dict(num_epochs=120),
    loss = "bcewithlogits",
    optimizer = "adam",


)
```

```python
entity_to_id = dataset.entity_to_id
id_to_entity = {value: key for key, value in entity_to_id.items()}

relation_to_id = dataset.relation_to_id
id_to_relation = {value: key for key, value in relation_to_id.items()}
```

```python
import requests

def fetch_label_from_wikidata(wikidata_id):
    # Prepare the Wikidata API URL
    url = f'https://www.wikidata.org/w/api.php?action=wbgetentities&ids={wikidata_id}&props=labels&format=json'

    # Send a GET request to the Wikidata API
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == requests.codes.ok:
        data = response.json()

        # Extract the label from the response
        try:
            label = data['entities'][wikidata_id]['labels']['en']['value']
            return label
        except KeyError:
            return f"No label found for entity ID {wikidata_id}"
    else:
        return "Request failed"
```

```python
import pandas as pd
import numpy as np

predictionsDF = pd.read_csv('first50Predictions.csv')

# Replace ids from dataset with their values in the dataframe
predictionsDF.head_id = [id_to_entity[id] for id in predictionsDF.head_id]
predictionsDF.relation_id = [id_to_relation[id] for id in predictionsDF.relation_id]
predictionsDF.tail_id = [id_to_entity[id] for id in predictionsDF.tail_id]

# Get all unique wikidate ids
wikidata_ids = list(set(predictionsDF.head_id.tolist() + predictionsDF.tail_id.tolist() + predictionsDF.relation_id.tolist()))

# Fetch all labels from wikidata and save them in a dictionary
wikidata_id_to_label = {}

for wikidata_id in wikidata_ids:
    label = fetch_label_from_wikidata(wikidata_id)
    wikidata_id_to_label[wikidata_id] = label

# Replace ids from wikidata with their labels in the dataframe
predictionsDF.head_id = [wikidata_id_to_label[id] for id in predictionsDF.head_id]
predictionsDF.relation_id = [wikidata_id_to_label[id] for id in predictionsDF.relation_id]
predictionsDF.tail_id = [wikidata_id_to_label[id] for id in predictionsDF.tail_id]

# Save the dataframe with the labels in a CSV
predictionsDF.to_csv("first50PredictionsLabels.csv", index = None)
```
